```python
#データの読み込み
import numpy as np
import pandas as pd
import math

data = pd.read_csv('titanic_train.csv')
df = pd.DataFrame(data)
df.head()
```

```python
#欠損値の有無の確認
df.isnull().sum()
```

```python
#敬称ごとの年齢の平均値を算出する
honer_ave = [0, 0, 0, 0]    #Mr, Mrs, Miss, Masterの順
honer_count = [0, 0, 0, 0]

df_ = df.dropna(subset=['Age'])

for i,name in enumerate(df_['Name']):
    if 'Mr.' in name:
        honer_ave[0] += df_.iloc[i, 5]
        honer_count[0] += 1
    elif 'Mrs.' in name:
        honer_ave[1] += df_.iloc[i, 5]
        honer_count[1] += 1
    elif 'Miss.' in name:
        honer_ave[2] += df_.iloc[i, 5]
        honer_count[2] += 1
    elif 'Master.' in name:
        honer_ave[3] += df_.iloc[i, 5]
        honer_count[3] += 1
        
for i in range(4):
    honer_ave[i] /= honer_count[i]

print(honer_ave)
```

```python
#Ageの欠損値を、敬称ごとの平均値をもとに補完する
for i, name in enumerate(df['Name']):
    if math.isnan(df.iloc[i, 5]):
        if 'Mr.' in name:
            df.iloc[i, 5] = honer_ave[0]
        elif 'Mrs.' in name:
            df.iloc[i, 5] = honer_ave[1]
        elif 'Miss.' in name:
            df.iloc[i, 5] = honer_ave[2]
        elif 'Master.' in name:
            df.iloc[i, 5] = honer_ave[3]

df.isnull().sum()
```

```python
#欠損値がまだ一つあるので確認する
for i, name in enumerate(df['Age']):
    if math.isnan(df.iloc[i, 5]):
        print(df.iloc[i, :])
```

```python
#名前にDr.があり、さらに男性であることから、Mr.と同じ扱いにする
df.iloc[766, 5] = honer_ave[0]
```
